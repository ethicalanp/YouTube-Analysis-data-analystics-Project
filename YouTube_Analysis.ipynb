```python
c=widgets.Dropdown(
    options=video["Genre"].unique(),
    value='Music',
    description='Number:',
    disabled=False,
)
def topchannel(a):
  channel = video[video["Genre"]==a]
  final = channel.groupby('Genre')[[ 'likes', 'dislikes']].sum().sort_values(by='likes', ascending=False).head(10)
  ib=["likes","dislikes"]
  val = [final["likes"].values[0],final['dislikes'].values[0]]
  plt.pie(val,labels=Ib,autopct='%1.1f%%', startangle=90)
  plt.title(a)
  plt.show()
widgets.interact(topchannel,a=c)
```
